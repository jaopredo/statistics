# Intervalo de Confiança de Wald
Usado para calcular intervalo de confiança da proporção de sucessos em uma população! Suponha que $X_1,...,X_n \sim Bern(p)$ i.i.d e temos uma amostra $x_1,...,x_n$. Para grandes valores de $n$, o intervalo de confiança de wald é uma boa medida, dado por:
$$
\lim_{n \rightarrow \infty} \mathbb{P}(p \in \left[ \hat{p} - z_{\frac{\alpha}{2}} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}, \hat{p} + z_{\frac{\alpha}{2}} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}} \right]) = 1 - \alpha
$$
onde $\hat{p}$ é a proporção da amostra, $z_{\frac{\alpha}{2}}$ e o $\frac{\alpha}{2}$ quantil da normal padrão e $n$ é o tamanho da amostra

```python
import numpy as np
from scipy.stats import bernoulli, norm
```

```python
p = 0.3 # Probabilidade de sucesso, na vida real é desconhecido
n = 10000 # Tamanho da amostra
alpha = 0.05 # Nível de significância
samples = bernoulli.rvs(p=p, size=n)
```

```python
positive_cases = np.sum(samples)
p_hat = positive_cases / n
```

```python
a = p_hat - norm.ppf(alpha/2, loc=0, scale=1) * np.sqrt(p_hat * (1 - p_hat) / n)
b = p_hat + norm.ppf(alpha/2, loc=0, scale=1) * np.sqrt(p_hat * (1 - p_hat) / n)
print(a,b)
```

Podemos ver que o intervalo de confiança **observado** chega muito próximo de 0.3. Vamos tirar várias amostras e calcular os intervalos observados e depois ver de quantos deles o meu p caiu

```python
k = 10000
contains_true_p = 0
for i in range(k):
    samples = bernoulli.rvs(p=p, size=n)
    positive_cases = np.sum(samples)
    p_hat = positive_cases / n
    a = p_hat + norm.ppf(alpha/2, loc=0, scale=1) * np.sqrt(p_hat * (1 - p_hat) / n)
    b = p_hat - norm.ppf(alpha/2, loc=0, scale=1) * np.sqrt(p_hat * (1 - p_hat) / n)
    if a <= p <= b:
        contains_true_p += 1

print(f"Proporção de intervalos que contêm o valor verdadeiro: {contains_true_p / k} que é aproximadamente 1-alpha")
```
