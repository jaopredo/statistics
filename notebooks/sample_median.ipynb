# Mediana Amostral
Em um exercício da lição 1, demonstramos um intervalo de confiança da mediana amostral de uma amostra qualquer. Vamos ver na prática! Intervalo:
$$
    \hat{M}_n \pm \frac{c}{2 f(\hat{M}_n) \sqrt{n}}
$$
onde $\hat{M}_n$ é a mediana amostral e $f(x)$ a densidade da distribuição

```python
import numpy as np
from scipy.stats import expon, norm
from math import factorial
```

Vamos tirar uma amostra de uma distribuição $Exponencial(\lambda)$, onde a mediana é dada por:
$$
    m = \frac{ln(2)}{\lambda}
$$

```python
lambd = 5  # taxa (lambda)
confidence = 0.95
n = 100
sample = np.random.exponential(scale=1 / lambd, size=n)
```

```python
sample_median = np.median(sample)
print(f"Mediana Amostral: {sample_median}")
```

```python
print(f"Mediana Real: {np.log(2)/lambd}")
```

```python
c = norm.ppf((1 + confidence) / 2)
```

```python
A = sample_median - c / (2 * np.sqrt(n) * expon.pdf(sample_median, scale=1 / lambd))
B = sample_median + c / (2 * np.sqrt(n) * expon.pdf(sample_median, scale=1 / lambd))

A, B
```

Esse intervalo, porém, pressupões amostras grandes para funcionar bem (Onde faz sentido o teorema central do limite). Então vou aplicar o outro intervalo que derivamos no último exercício da lição 1

```python
n = 15
sample = np.random.exponential(scale=1 / lambd, size=n)
ordered_sample = np.sort(sample)
conficence = 0.95
```

```python
def n_chooses_k(n, k):
    return factorial(n) // (factorial(k) * factorial(n - k))
```

```python
j = 0
cum_sum = 0
while j < n:
    cum_sum = n_chooses_k(n, j)
    if cum_sum/(2**n) >= confidence:
        break
    j += 1
```

```python
print(f"Intervalo de confiança: [{ordered_sample[n-j]}, {ordered_sample[n-1]}]")
```
